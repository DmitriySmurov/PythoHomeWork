```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Bitcoin BEP2.csv')
df2 = pd.read_csv('ethereum.csv')
```

```python
def plot_btc():
    
    plt.figure(figsize=(25,10))

    
    btc_close = df1['Close']
    btc_open = df1['Open']

   
    plt.plot(btc_close, label='Closing', color='red')
    plt.plot(btc_open, label='Open', color='green')
    plt.legend()
   
```

```python
def plot_eth():
    
    plt.figure(figsize=(25,10))

  
    eth_close = df2['Close']
    eth_open = df2['Open']

   
    plt.plot(eth_close, label='Closing', color='red')
    plt.plot(eth_open, label='Open', color='green')
    plt.legend()
  
```

```python
plot_btc()
```

```python
plot_eth()
```
